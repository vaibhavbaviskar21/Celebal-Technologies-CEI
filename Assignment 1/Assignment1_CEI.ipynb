Identifying missing values

```python
print(df.isnull().sum())
```

Filling missing discount values with 0

```python
if 'discount' in df.columns:
    df['discount'] = df['discount'].fillna(0)
```

Defining columns to drop

```python
columns_to_drop = [
    'videos',
    'variations',
    'what_customers_said',
    'seller_name',
    'seller_information'
]
```

Dropping columns with excessive missing data

```python
df.drop(columns=columns_to_drop, inplace=True, errors='ignore')
```

Missing values after handling

```python
print(df.isnull().sum())
```

Converting final_price to numeric format

```python
if 'final_price' in df.columns:
    df['final_price'] = (
        df['final_price']
        .astype(str)
        .str.replace('₹', '', regex=False)
        .str.replace(',', '', regex=False)
        .str.replace('"', '', regex=False)
    )
    df['final_price'] = pd.to_numeric(
        df['final_price'],
        errors='coerce'
    )
    df.dropna(subset=['final_price'], inplace=True)
```

Displaying final_price data type

```python
print(df['final_price'].dtype)
```

Selecting specific columns

```python
selected_columns = [
    col for col in ['title', 'rating', 'final_price']
    if col in df.columns
]
display(df[selected_columns].head())
```

Filtering products with rating greater than 0

```python
if 'rating' in df.columns:
    filtered_df = df[df['rating'] > 0]
    print("Filtered Shape:", filtered_df.shape)
    df = filtered_df
    display(filtered_df.head())
```

Removing duplicate product entries

```python
rows_before = len(df)
if 'product_id' in df.columns:
    df.drop_duplicates(subset=['product_id'], inplace=True)
else:
    df.drop_duplicates(inplace=True)
rows_after = len(df)
print("Rows Before:", rows_before)
print("Rows After :", rows_after)
print("Duplicates Removed:", rows_before - rows_after)
```

Setting random seed for quantity generation

```python
np.random.seed(42)
```

Generating a sample quantity column

```python
df['quantity'] = np.random.randint(
    1,
    6,
    size=len(df)
)
```

Calculating total amount

```python
df['total_amount'] = (
    df['final_price']
    * df['quantity']
)
```

Displaying derived columns

```python
display(
    df[
        ['final_price',
         'quantity',
         'total_amount']
    ].head()
)
```
